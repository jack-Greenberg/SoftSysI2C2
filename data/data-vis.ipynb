```python
scl = pd.read_csv('scl_oscilliscope.csv')
time = scl['Time (s)']
sig = scl['Channel 1 (V)']

fig, ax = plt.subplots()

plt.title('SCL signal at 100kHz', {'fontsize': 16})
ax.plot([t * 10**6 for t in time], sig)
ax.set_xlim(-30, 30)
ax.grid(True)
ax.set_xlabel('Time (μs)')
ax.set_ylabel('Voltage (V)')

fig.tight_layout()
plt.savefig('scl.png', dpi=300)
plt.show()
```

```python
start = pd.read_csv('start_i2c-stop_i2c.csv')
time = [ (t - start['Time (s)'][0]) * 10**6 for t in start['Time (s)']]
sda = start['Channel 1 (V)']
scl = start['Channel 2 (V)']

fig, axs = plt.subplots(2, 1)
fig.suptitle('Start Condition', fontsize=16)
fig.subplots_adjust(hspace=0.5)

START = 190
STOP = 240
axs[0].plot(time, sda)
axs[0].set_xlim(START, STOP)
axs[0].set_xlabel('Time (μs)')
axs[0].set_ylabel('SDA')

axs[1].plot(time, scl)
axs[1].set_xlim(START, STOP)
axs[1].set_xlabel('Time (μs)')
axs[1].set_ylabel('SCL')

# fig.tight_layout()
plt.savefig('start_condition.png', dpi=300)
plt.show()
```
